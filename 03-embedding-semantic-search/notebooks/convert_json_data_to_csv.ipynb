```python
import pandas as pd
import json

# Define paths
json_path = '/content/drive/MyDrive/nawy /locations_compounds_and_blogs_data.json'
csv_path = '/content/drive/MyDrive/nawy /locations_compounds_and_blogs_data.csv'

# Load the JSON data
with open(json_path, 'r') as f:
    data = json.load(f)

# Convert to DataFrame
# Note: This assumes the JSON is a list of objects or a standard structure pandas can parse
df = pd.json_normalize(data)

# Save to CSV
df.to_csv(csv_path, index=False)

print(f'Successfully converted JSON to CSV: {csv_path}')
df.head()
```
