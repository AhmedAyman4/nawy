```python
import re
import json
import os

def convert_text_to_json(input_text):
    """
    Parses the text file based on URL markers and extracts content.
    """
    # Split the text by the URL marker blocks
    # This looks for the repeated equal signs followed by URL:
    chunks = re.split(r'={10,}\s+URL:\s+', input_text)

    json_data = []

    for chunk in chunks:
        if not chunk.strip():
            continue

        # Extract the URL (everything until the next newline or equal sign)
        url_match = re.match(r'(https?://[^\s\n=]+)', chunk)
        if not url_match:
            continue

        url = url_match.group(1)

        # The content is everything after the URL block
        # We remove the trailing separator line from the URL block
        content = re.sub(r'^.*?\s+={10,}', '', chunk, count=1, flags=re.DOTALL).strip()

        # Clean up any trailing equal sign separators at the very end of the content
        content = re.sub(r'={10,}$', '', content).strip()

        json_data.append({
            "url": url,
            "content": content
        })

    return json_data

def main():
    # Path provided by the user
    input_file_path = "/content/drive/MyDrive/nawy /locations_compounds_and_blogs_data.txt"
    # Updated output name as requested
    output_file_path = "locations_compounds_and_blogs_data.json"

    if not os.path.exists(input_file_path):
        print(f"Error: File not found at {input_file_path}")
        return

    print(f"Reading file from {input_file_path}...")
    with open(input_file_path, 'r', encoding='utf-8') as f:
        raw_text = f.read()

    print("Converting text to JSON objects...")
    processed_data = convert_text_to_json(raw_text)

    print(f"Saving {len(processed_data)} objects to {output_file_path}...")
    with open(output_file_path, 'w', encoding='utf-8') as f:
        json.dump(processed_data, f, indent=4, ensure_ascii=False)

    print(f"Success! You can now use '{output_file_path}' for your RAG system.")

if __name__ == "__main__":
    main()
```
